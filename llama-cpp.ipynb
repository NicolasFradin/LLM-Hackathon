```python
import os
import urllib.request


def download_file(file_link, filename):
    # Checks if the file already exists before downloading
    if not os.path.isfile(filename):
        urllib.request.urlretrieve(file_link, filename)
        print("File downloaded successfully.")
    else:
        print("File already exists.")

# Dowloading GGML model from HuggingFace
ggml_model_path = "https://huggingface.co/TheBloke/Llama-2-70B-GGUF/blob/main/llama-2-70b.Q4_K_M.gguf"
filename = "llama-2-70b.Q4_K_M.gguf"

download_file(ggml_model_path, filename)
```

```python
from llama_cpp import Llama

filename = "codellama-7b-python.Q4_K_M.gguf"
llm = Llama(model_path=filename, n_ctx=1024, n_batch=126)
```

```python
def generate_text(
    prompt="Who is the CEO of Apple?",
    max_tokens=256,
    temperature=0.8,
    top_p=0.5,
    echo=False,              # Echo the prompt back in the output
    stop=["#"],  # Stop generating just before the model would generate a new question
):
    output = llm(
        prompt,
        max_tokens=max_tokens,
        temperature=temperature,
        top_p=top_p,
        echo=echo,
        stop=stop,
    )
    output_text = output["choices"][0]["text"].strip()
    return output_text


def generate_prompt_from_template(input):
    #chat_prompt_template = f"""<im_start|>system
    #    You are a helpful chatbot.<|im_end|>
    #    <|im_start|>user
    #    {input}<|im_end|>"""

    #You are a helpful assistant. You do not respond as 'User' or pretend to be 'User'. You only respond once as 'Assistant'.
    #input = "You are a car seller. You can't lie or say something false. Don't include phone number. Create a more textual description based on the following details of the car:" + str(input) + ". Don't tell anything about the owner's phone number or any private information. Limit your response at 100 words in an only paragraphe."
    input = "Don't add anything new and write a json with this type: {'query': , 'filters': }"
    chat_prompt_template = f"""User: {input}
                               Assistant:"""
    return chat_prompt_template


prompt = generate_prompt_from_template(
    example
)

print('prompt:', prompt)

for i in range(5):
    text = generate_text(
        prompt,
        #max_tokens=100,
    )
    print(text)
    print()
```
